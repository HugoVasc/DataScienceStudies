```python
from collections import Counter
import numpy as np

class KNN():
    def __init__(self, k):
        self.k = k
    
    def fit(self, X, y):
        self.X = X
        self.y = y

    def _distance(self, x1, x2):
        return np.sqrt(np.sum((x1 - x2) ** 2))

    def _predict_one(self, x):
        print('Entrei na função predict_one')
        distances = []
        for j in range(self.X.shape[0]):
            print(f'Entrei no for (predict_one) pela {j}ª vez')
            distances.append((j, self._distance(x, self.X[j])))
        print('Sai do for (predict_one)')
        distances.sort(key=lambda x: x[1])
        top_k = [self.y[i] for i, _ in distances[:self.k]]
        return Counter(top_k).most_common(1)[0][0]

    def predict(self, X):
        y_pred = []
        print('Entrei na função predict')
        for i in range(X.shape[0]):
            print(f'Entrei no for (predict) pela {i}ª vez')
            y_pred.append(self._predict_one(X[i]))
        print('Sai da função predict')
        return np.array(y_pred)
    
```

Separando as labels do dataset

```python
class_is_red = df.wine_is_red
class_quality = df.quality
X = df.drop('wine_is_red', axis=1)
X = X.drop('quality', axis=1)
print(X.shape, class_is_red.shape, class_quality.shape, sep='\n')
```
